```python
plik = pd.ExcelFile('Livechat.xlsx') #wczytanie pliku .xls
df_temp = plik.parse() #zamiana tabeli excel na ramkę danych
df_temp
```

```python
df_temp=df_temp.drop(columns=['Temp1','Temp2','Temp3','Temp4','Temp5','Temp6','Temp7'], axis=1)#usunięcie niepotrzebnych kolumn
df=df_temp.drop(0, axis=0)
df
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['w tys. PLN'],df['Przychody ze sprzedaży'], color = 'gray', alpha=0.5, label = 'Przychody w tys. PLN')
plt.bar(df['w tys. PLN'],df['Zysk ze sprzedaży'], color = 'darkgray', alpha=0.5, label = 'Zysk ze sprzedaży w tys. PLN')
plt.xticks(df['w tys. PLN'])
 
plt.legend(frameon=False)
plt.xlabel('Lata')
plt.ylabel('Wartość')
plt.title('Przychody spółki');
```

# Wskaźniki

```python
#Rentowność
df['ROE %'] = round((df['Zysk netto']/df['Kapitał własny akcjonariuszy jednostki dominującej'])*100,2) #najlepsze >20%
df['ROA %'] = round((df['Zysk netto']/df['Aktywa razem'])*100,2) #najlepsze >10%
df['ROS %'] = round((df['Zysk ze sprzedaży']/df['Przychody ze sprzedaży'])*100,2) #najlepsze >10%
df['Rentowność operacyjna %'] = round((df['Zysk operacyjny (EBIT)']/df['Przychody ze sprzedaży'])*100,2) #najlepsza >10%
df['Rentowność zysku netto %'] = round((df['Zysk netto']/df['Przychody ze sprzedaży'])*100,2) #najlepsza >10%

#Płynność
df['Płynność bieżąca'] = round((df['Aktywa obrotowe']/df['Zobowiązania krótkoterminowe']),2) #najlepsza >2
df['Płynność gotówkowa %'] = round((df['Środki pieniężne i inne aktywa pieniężne']/df['Zobowiązania krótkoterminowe'])*100,2) #najlepsza >20%
df['Kapital obrotowy'] = round((df['Aktywa obrotowe']-df['Zobowiązania krótkoterminowe']),2) #najlepszy >0

#Aktywności
```
