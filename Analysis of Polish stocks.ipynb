# Indicators 

```python
#Book value 
for i in range(2, df.Year.count()+1):
    #df.loc[1,'Change in book value %']=100
    df.loc[i,'Change in book value %']=round(((df.loc[i,'Equity']- \
                                         df.loc[i-1,'Equity']) \
                                         /df.loc[i-1,'Equity'])*100,0) #the best >0
#Increasing sales revenues
for i in range(2, df.Year.count()+1):
    #df.loc[1,'Change in sales revenues %']=100
    df.loc[i,'Change in revenues %']=round(((df.loc[i,'Revenue']- \
                                                   df.loc[i-1,'Revenue']) \
                                                   /df.loc[i-1,'Revenue'])*100,0) #the best >0

#Increasing net income
for i in range(2, df.Year.count()+1):
    #df.loc[1,'Change in net income %']=100
    df.loc[i,'Change in net income %']=round(((df.loc[i,'Net income']-df.loc[i-1,'Net income'])\
                                              /df.loc[i-1,'Net income'])*100,0) #the best >0
```

```python
#Profitability 
df['ROE %'] = round((df['Net income']/df['Equity'])*100,2) #the best >20%
df['ROA %'] = round((df['Net income']/df['Total assets'])*100,2) #the best >10%
df['ROS %'] = round((df['Gross Profit']/df['Revenue'])*100,2) #the best >10%
df['Operating margin %'] = round((df['Operating income (EBIT)']/df['Revenue'])*100,2) #the best >10%
df['Net margin %'] = round((df['Net income']/df['Revenue'])*100,2) #the best >10%
```

```python
#Liquidity 
df['Current ratio'] = round((df['Current assets']/df['Short term debt']),2) #the best >2
df['Cash ratio %'] = round((df['Cash and equivalents']/df['Short term debt'])*100,2) #the best >20%
df['Working capital'] = round((df['Current assets']-df['Short term debt']),0) #the best >0
```

```python
#Activities 
df['Days sales outstanding'] = round((df['Short term debt']*365/df['Revenue']),0) #
df['Days inventory'] = round((df['Inventory']*365/df['Revenue']),0) #
df['Payables period'] = round((df['Short term debt']*365/df['Revenue']),0) #
df['Cash conversion cycle'] = round((df['Days sales outstanding']+df['Days inventory']-df['Payables period']),0) #the best <0
```

```python
#Liabilities 
df['Longterm Debt/Equity Ratio'] = round((df['Long term debt'] \
                                    /df['Equity'])*100,2) #the best <20%
df['Debt ratio'] = round(((df['Long term debt']+df['Short term debt']) \
                          /df['Total assets'])*100,2) #the best <67%
```

```python
#Property 
df['1st degree of coverage'] = round((df['Equity']/df['Fixed assets']),2) #the best >0.6
```

```python
#Cash flow
def flow(data):
        if data > 0: return 1
        else: return 0
        
df['Operating cash flow rate']=0        
df['Operating fash flow rate']=df['Operating cash flow'].apply(flow) #the best >0

df['Investing cash flow rate']=0        
df['Investing cash flow rate']=df['Investing cash flow'].apply(flow) #the best <0

df['Finance cash flow rate']=0        
df['Finance cash flow rate']=df['Finance cash flow'].apply(flow) #the best <0

df['Free cash flow rate']=0        
df['Free cash flow rate']=df['Free cash flow'].apply(flow) #the best >0
```

```python
#Cash flow description

def flows(data):
        if ((data['Operating cash flow rate'] > 0) & \
        (data['Investing cash flow rate'] > 0) & \
        (data['Finance cash flow rate'] > 0)) : return 'I - (+++) rare situation, need to analyze cases'      
        elif ((data['Operating cash flow rate'] > 0) & \
        (data['Investing cash flow rate']) > 0 & \
        (data['Finance cash flow rate']) == 0): return 'II - (++-) restructuring process'    
        elif ((data['Operating cash flow rate'] > 0) & \
        (data['Investing cash flow rate']) == 0 & \
        (data['Finance cash flow rate']) == 0): return 'III - (+--) stable company, the best'    
        elif ((data['Operating cash flow rate'] == 0) & \
        (data['Investing cash flow rate']) == 0 & \
        (data['Finance cash flow rate']) == 0) : return 'IV - (---) bankrupt'    
        elif ((data['Operating cash flow rate'] < 0) & \
        (data['Investing cash flow rate']) > 0 & \
        (data['Finance cash flow rate']) > 0)  : return 'V - (-++) temporary problems'    
        elif ((data['Operating cash flow rate'] == 0) & \
        (data['Investing cash flow rate']) == 0 & \
        (data['Finance cash flow rate']) > 0)  : return 'VI - (--+) start-up, early stage of activity'    
        elif ((data['Operating cash flow rate'] == 0) & \
        (data['Investing cash flow rate']) > 0 & \
        (data['Finance cash flow rate']) == 0)   : return 'VII - (-+-) probable bankruptcy'    
        elif ((data['Operating cash flow rate'] > 0) & \
        (data['Investing cash flow rate']) == 0 & \
        (data['Finance cash flow rate']) == 0)  : return 'VIII - (+-+) after the start-up phase'                                 
        else: return 'Error'

df['Cash flow description']='0'      
df['Cash flow description']=df.apply(flows, axis=1)
```

## The rating of the company

```python
#Score based on fundamental indicators

def scores(data):
    l=0
    
    #Dyvidend
    if (data['Dividend rate'] > 2): l=l+1;
    if (data['Dividend'] > 1): l=l+1;
        
    #Book value
    if (data['Change in book value %'] > 0): l=l+1;
    
    #Increasing sales revenues 
    if (data['Change in revenues %'] > 0): l=l+1;
    
    #Increasing net profit 
    if (data['Change in net income %'] > 0): l=l+1;
        
    #Profitability
    if (data['ROE %'] > 20): l=l+1;
    if (data['ROA %'] > 10): l=l+1;
    if (data['ROS %'] > 10): l=l+1;
    if (data['Operating margin %'] > 10): l=l+1;
    if (data['Net margin %'] > 10): l=l+1;
    
    #Liquidity    
    if (data['Current ratio'] > 2): l=l+1;
    if (data['Cash ratio %'] > 20): l=l+1;
    if (data['Working capital'] > 0): l=l+1;
    
    #Activity    
    if (data['Cash conversion cycle'] < 0): l=l+1;
    
    #Debt  
    if (data['Longterm Debt/Equity Ratio'] < 20): l=l+1;
    if (data['Debt ratio'] < 67): l=l+1;
        
    #Cash flow description
    if (data['Cash flow description'] == 'III - (+--) stable company, the best'): l=l+1    
    elif (data['Cash flow description'] == 'VI - (--+) start-up, early stage of activity'): l=l+1
    elif (data['Cash flow description'] == 'VIII - (+-+) after the start-up phase'): l=l+1;
    
    #Property
    if (data['1st degree of coverage'] > 0.6): l=l+1;
        
    return l
        
df['Scores']='0'      
df['Scores']=df.apply(scores, axis=1)            
```

Rating:
- 15-18: Very good company
- 12-14: Good company
- 9-11: Average company
- 0-9: Weak company

```python
#Final evaluation

def rate(score):
    if (score >= 15): return ('Very good company')
    elif (score >= 12): return ('Good company')
    elif (score >= 9): return ('Average company')
    else: return ('Weak company')

df['Rating'] = df['Scores'].apply(rate);
```
