# Wskaźniki

```python
#Wartość księgowa
for i in range(2,df.Lata.count()+1):
    df.loc[1,'Zmiana wartości księgowej %']=100
    df.loc[i,'Zmiana wartości księgowej %']=round(((df.loc[i,'Kapitał własny akcjonariuszy jednostki dominującej']- \
                                         df.loc[i-1,'Kapitał własny akcjonariuszy jednostki dominującej']) \
                                         /df.loc[i-1,'Kapitał własny akcjonariuszy jednostki dominującej'])*100,0) #najlepsze >0
#Wzrastanie przychodów ze sprzedaży
for i in range(2,df.Lata.count()+1):
    df.loc[1,'Zmiana przychodów ze sprzedaży %']=100
    df.loc[i,'Zmiana przychodów ze sprzedaży %']=round(((df.loc[i,'Przychody ze sprzedaży']- \
                                                           df.loc[i-1,'Przychody ze sprzedaży']) \
                                                           /df.loc[i-1,'Przychody ze sprzedaży'])*100,0) #najlepsze >0

#Wzrastanie zysku netto
for i in range(2,df.Lata.count()+1):
    df.loc[1,'Zmiana zysku netto %']=100
    df.loc[i,'Zmiana zysku netto %']=round(((df.loc[i,'Zysk netto']-df.loc[i-1,'Zysk netto'])\
                                              /df.loc[i-1,'Zysk netto'])*100,0) #najlepsze >0
```

```python
#Rentowność
df['ROE %'] = round((df['Zysk netto']/df['Kapitał własny akcjonariuszy jednostki dominującej'])*100,2) #najlepsze >20%
df['ROA %'] = round((df['Zysk netto']/df['Aktywa razem'])*100,2) #najlepsze >10%
df['ROS %'] = round((df['Zysk ze sprzedaży']/df['Przychody ze sprzedaży'])*100,2) #najlepsze >10%
df['Rentowność operacyjna %'] = round((df['Zysk operacyjny (EBIT)']/df['Przychody ze sprzedaży'])*100,2) #najlepsza >10%
df['Rentowność zysku netto %'] = round((df['Zysk netto']/df['Przychody ze sprzedaży'])*100,2) #najlepsza >10%
```

```python
#Płynność
df['Płynność bieżąca'] = round((df['Aktywa obrotowe']/df['Zobowiązania krótkoterminowe']),2) #najlepsza >2
df['Płynność gotówkowa %'] = round((df['Środki pieniężne i inne aktywa pieniężne']/df['Zobowiązania krótkoterminowe'])*100,2) #najlepsza >20%
df['Kapital obrotowy'] = round((df['Aktywa obrotowe']-df['Zobowiązania krótkoterminowe']),0) #najlepszy >0
```

```python
#Aktywności
df['Cykl inkasa należności'] = round((df['Należności krótkoterminowe']*365/df['Przychody ze sprzedaży']),0) #
df['Rotacja zapasów'] = round((df['Zapasy']*365/df['Przychody ze sprzedaży']),0) #
df['Cykl spłaty zobowiązań'] = round((df['Zobowiązania krótkoterminowe']*365/df['Przychody ze sprzedaży']),0) #
df['Cykl konwersji gotówki'] = round((df['Cykl inkasa należności']+df['Rotacja zapasów']-df['Cykl spłaty zobowiązań']),0) #najlepszy <0
```

```python
#Zadłużenie
df['Zadłużenie długoterminowe kapitału własnego'] = round((df['Zobowiązania długoterminowe'] \
                                                          /df['Kapitał własny akcjonariuszy jednostki dominującej'])*100,2) #najlepsza <20%
df['Wskaźnik ogólnego zadłużenia'] = round(((df['Zobowiązania długoterminowe']+df['Zobowiązania krótkoterminowe']) \
                                            /df['Aktywa razem'])*100,2) #najlepsza <67%
```

```python
#Majątkowe
df['I stopień pokrycia'] = round((df['Kapitał własny akcjonariuszy jednostki dominującej']/df['Aktywa trwałe']),2) #najlepsza >0.6
```

```python
#Przepływy pieniężne
def przepływ(typ):
        if typ > 0: return 1
        else: return 0
        
df['Przepływy pieniężne z działalności operacyjnej ocena']=0        
df['Przepływy pieniężne z działalności operacyjnej ocena']=df['Przepływy pieniężne z działalności operacyjnej'].apply(przepływ) #najlepsze >0

df['Przepływy pieniężne z działalności inwestycyjnej ocena']=0        
df['Przepływy pieniężne z działalności inwestycyjnej ocena']=df['Przepływy pieniężne z działalności inwestycyjnej'].apply(przepływ) #najlepsze <0

df['Przepływy pieniężne z działalności finansowej ocena']=0        
df['Przepływy pieniężne z działalności finansowej ocena']=df['Przepływy pieniężne z działalności finansowej'].apply(przepływ) #najlepsze <0

df['Przepływy pieniężne razem ocena']=0        
df['Przepływy pieniężne razem ocena']=df['Przepływy pieniężne razem'].apply(przepływ) #najlepsze >0
```

```python
#Przepływy pieniężne opis

def przepływy(typ):
        if ((typ['Przepływy pieniężne z działalności operacyjnej ocena'] > 0) & \
        (typ['Przepływy pieniężne z działalności inwestycyjnej ocena'] > 0) & \
        (typ['Przepływy pieniężne z działalności finansowej ocena'] > 0)) : return 'I - (+++) rzadka sytuacja, potrzeba analizy sprawzd.'      
        elif ((typ['Przepływy pieniężne z działalności operacyjnej ocena'] > 0) & \
        (typ['Przepływy pieniężne z działalności inwestycyjnej ocena']) > 0 & \
        (typ['Przepływy pieniężne z działalności finansowej ocena']) == 0): return 'II - (++-) proces restrukturyzacji'    
        elif ((typ['Przepływy pieniężne z działalności operacyjnej ocena'] > 0) & \
        (typ['Przepływy pieniężne z działalności inwestycyjnej ocena']) == 0 & \
        (typ['Przepływy pieniężne z działalności finansowej ocena']) == 0): return 'III - (+--) stabilna spółka, najlepsza sytuacja'    
        elif ((typ['Przepływy pieniężne z działalności operacyjnej ocena'] == 0) & \
        (typ['Przepływy pieniężne z działalności inwestycyjnej ocena']) == 0 & \
        (typ['Przepływy pieniężne z działalności finansowej ocena']) == 0) : return 'IV - (---) bankrut'    
        elif ((typ['Przepływy pieniężne z działalności operacyjnej ocena'] < 0) & \
        (typ['Przepływy pieniężne z działalności inwestycyjnej ocena']) > 0 & \
        (typ['Przepływy pieniężne z działalności finansowej ocena']) > 0)  : return 'V - (-++) tymczasowe problemy'    
        elif ((typ['Przepływy pieniężne z działalności operacyjnej ocena'] == 0) & \
        (typ['Przepływy pieniężne z działalności inwestycyjnej ocena']) == 0 & \
        (typ['Przepływy pieniężne z działalności finansowej ocena']) > 0)  : return 'VI - (--+) start-up, wczesna faza dzialalności'    
        elif ((typ['Przepływy pieniężne z działalności operacyjnej ocena'] == 0) & \
        (typ['Przepływy pieniężne z działalności inwestycyjnej ocena']) > 0 & \
        (typ['Przepływy pieniężne z działalności finansowej ocena']) == 0)   : return 'VII - (-+-) prawdopodobny bankrut'    
        elif ((typ['Przepływy pieniężne z działalności operacyjnej ocena'] > 0) & \
        (typ['Przepływy pieniężne z działalności inwestycyjnej ocena']) == 0 & \
        (typ['Przepływy pieniężne z działalności finansowej ocena']) == 0)  : return 'VIII - (+-+) po fazie start-up'                                 
        else: return 'Błąd'

df['Przepływy pieniężne opis']='0'      
df['Przepływy pieniężne opis']=df.apply(przepływy, axis=1)
```

## Ocena spółki

```python
#Ocena punktowa na podstawie wskaźników fundamentalnych

def punkty(dane):
    l=0
    
    #Dywidenda
    if (dane['Stopa dywidendy'] > 2): l=l+1;
    if (dane['Wartość dywidendy na akcje'] > 1): l=l+1;
        
    #Wartość księgowa
    if (dane['Zmiana wartości księgowej %'] > 0): l=l+1;
    #Wzrastanie przychodów ze sprzedaży
    if (dane['Zmiana przychodów ze sprzedaży %'] > 0): l=l+1;
    #Wzrastanie zysku netto
    if (dane['Zmiana zysku netto %'] > 0): l=l+1;
        
    #Rentowność
    if (dane['ROE %'] > 20): l=l+1;
    if (dane['ROA %'] > 10): l=l+1;
    if (dane['ROS %'] > 10): l=l+1;
    if (dane['Rentowność operacyjna %'] > 10): l=l+1;
    if (dane['Rentowność zysku netto %'] > 10): l=l+1;
    #Płynność    
    if (dane['Płynność bieżąca'] > 2): l=l+1;
    if (dane['Płynność gotówkowa %'] > 20): l=l+1;
    if (dane['Kapital obrotowy'] > 0): l=l+1;
    #Aktywności    
    if (dane['Cykl konwersji gotówki'] < 0): l=l+1;
    #Zadłużenie  
    if (dane['Zadłużenie długoterminowe kapitału własnego'] < 20): l=l+1;
    if (dane['Wskaźnik ogólnego zadłużenia'] < 67): l=l+1;
        
    #Przepływy pieniężne opis
    if (dane['Przepływy pieniężne opis'] == 'III - (+--) stabilna spółka, najlepsza sytuacja'): l=l+1    
    elif (dane['Przepływy pieniężne opis'] == 'VI - (--+) start-up, wczesna faza dzialalności'): l=l+1
    elif (dane['Przepływy pieniężne opis'] == 'VIII - (+-+) po fazie start-up'): l=l+1;
    #Majątkowe
    if (dane['I stopień pokrycia'] > 0.6): l=l+1;
        
    return l
        
df['Punkty']='0'      
df['Punkty']=df.apply(punkty, axis=1)            
```

```python
#Ocena:
#15-18: Bardzo dobra spółka
#12-14: Dobra spółka
#9-11: Śrenia spółka
#0-9: Słaba spółka

def ocena(punkt):
    if (punkt >= 15): return ('Bardzo dobra spółka')
    elif (punkt >= 12): return ('Dobra spółka')
    elif (punkt >= 9): return ('Średnia spółka')
    else: return ('Słaba spółka')

df['Ocena'] = df['Punkty'].apply(ocena)
```
